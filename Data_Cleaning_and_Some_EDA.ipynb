# Data Cleaning

##Change column dtypes to appropriate dtypes

Columns to change:
- LAST EDITED DATE (obj -> datetime)
- OBSERVATION COUNT (obj - int)
- OBSERVATION DATE (obj -> datetime)
- TIME OBSERVATION STARTED (obj -> time)


```python
df_clean = df.copy()
```

```python
#change X's in OBSERVATION COUNT to 1's
df_clean['OBSERVATION COUNT'].replace(to_replace=dict(X=1), inplace=True)
```

```python
df_clean = df_clean.astype({'OBSERVATION COUNT':'int'})
```

```python
df_clean['LAST EDITED DATE'] = pd.to_datetime(df_clean['LAST EDITED DATE'])
df_clean['OBSERVATION DATE'] = pd.to_datetime(df_clean['OBSERVATION DATE'])
df_clean['TIME OBSERVATIONS STARTED'] = pd.to_datetime(df_clean['TIME OBSERVATIONS STARTED'])
```

```python
df_clean.dtypes
```

## Clean out the states that are not New Mexico

```python
df_clean['STATE'].value_counts()
```

```python
filter = df_clean['STATE CODE'] != 'US-NM'
```

```python
df_clean.drop(index = df_clean[filter].index, inplace = True )
```

```python
df_clean.shape
```

## Turn our normal dataframe to a geopandas dataframe

```python
gdf = gpd.GeoDataFrame(
    df_clean, geometry=gpd.points_from_xy(df_clean.LONGITUDE, df_clean.LATITUDE), crs="EPSG:4326"
)
```

```python
gdf.geometry
```
